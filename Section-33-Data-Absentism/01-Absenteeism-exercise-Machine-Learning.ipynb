### Creating a logistic regression to predict absenteeism

#### Import the relevant libraries

```python
import pandas as pd
import numpy as np
```

#### Load the data

```python
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
```

```python
data_preprocessed.head()
```

#### Create the targets

```python
median_absenteeism_hours = data_preprocessed['Absenteeism Time in Hours'].median()
median_absenteeism_hours
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > median_absenteeism_hours, 1, 0) # another way of mapping data
```

```python
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
data_preprocessed.head()
```

#### A comment on the targets

We have implicitly balanced the data set by mapping it in two classes.

This will prevent our model to predict only 0 or only 1. (Not preference)

This is easly proved by dividing the number of targets that are one by the total number of targets.

```python
targets.sum() / targets.shape[0] # check that the targets are balanced between 45-55% of data in one side with respect to other side.
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours'], axis=1)
```
